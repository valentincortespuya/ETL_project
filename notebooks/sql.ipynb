```python
for index, row in matriculas.iterrows():
    try:
        query = f"""
        INSERT INTO matriculas (año, genero, universidadtipo, formato, facultad, matriculas, idmatriculas)
        VALUES ('{row["año"]}', '{row["genero"]}', '{row["universidad_tipo"]}', '{row["formato"]}', '{row["facultad"]}','{row["matriculas"]}', '{row["idmatriculas"]}');"""
        
        print(query)
        engine.execute(query)
    except:
        print(row["año"])
```
