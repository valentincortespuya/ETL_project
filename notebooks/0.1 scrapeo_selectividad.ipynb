# función para hacer iterrows para todos los dt anteriores!!

```python
import pandas as pd

def procesar_dataframes(lista_dataframes):
    
    datos = []

    
    for df in lista_dataframes:
        
        
        nuevos_nombres = df.columns.get_level_values(0).tolist()

         
        df.columns = nuevos_nombres

        
        
        
        for indice, fila in df.iterrows():
            
            facultad = df.columns[0]

            
            grado = fila[0]

            
            for i in range(1, len(fila), 1):
               
                año = df.columns[i]

                
                valor_selectividad = fila[i]

                
                datos.append([facultad, grado, valor_selectividad, año])

    
    df_nuevo = pd.DataFrame(datos, columns=['facultad', 'grado', 'valor selectividad', 'año'])

    
    return df_nuevo
```

```python

lista_dataframes = [df_selec_art_hum, df_selec_ing_arc, df_selec_ciencias_sociales, df_selec_ciencias_salud, df_selec_ciencias]


df_final = procesar_dataframes(lista_dataframes)


df_final
```

```python
df_final['valor selectividad'] = df_final['valor selectividad'].astype(str)
df_final['valor selectividad'] = df_final['valor selectividad'].apply(lambda x: re.sub(r'\(.*?\)|\*', '', x))
df_final['valor selectividad'] = df_final['valor selectividad'].str.replace(',', '.')
df_final['año'] = df_final['año'].str.replace('/', '-20')


df_final


```

```python
dt_selectividad_murcia = df_final.copy()

dt_selectividad_murcia.to_csv("../DATA/selectividad/dt_selectividad_murcia.csv", index=False)
```
