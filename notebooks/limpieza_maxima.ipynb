```python
df_matriculas_murcia_esp = pd.read_csv('datos_murcia_esp.csv',sep=',')
df_matriculas_murcia_esp.head(100
                )
```

```python
df_matriculas_murcia_esp = df_matriculas_murcia_esp.drop(df_matriculas_murcia_esp.columns[:7], axis=1)
```

```python



# Obtener las columnas del DataFrame original excepto las últimas 3 columnas
columnas = df_matriculas_murcia_esp.columns[:-3]

# Crear una lista para almacenar los elementos y los años
elementos = []
años = []

# Iterar sobre las filas del DataFrame original
for _, row in df_matriculas_murcia_esp.iterrows():
    # Iterar sobre las columnas seleccionadas
    for col in columnas:
        # Obtener el dato de la columna actual y el año correspondiente
        dato = row[col]
        año = col[-9:]  # Extraer el año de la columna
        
        # Agregar el dato y el año a las listas
        elementos.append(dato)
        años.append(año)

# Crear el nuevo DataFrame con las columnas "Datos" y "Año"
df_nuevo_murcia = pd.DataFrame({'Datos': elementos, 'Año': años})

# Mostrar el nuevo DataFrame




```

```python


import pandas as pd
import numpy as np

# Obtener las últimas tres columnas del DataFrame original
columnas_ultimas = df_matriculas_murcia_esp.columns[-3:]

# Crear el nuevo DataFrame con las repeticiones
df_nuevo_murcia2 = pd.DataFrame()
for index, row in df_matriculas_murcia_esp.iterrows():
    datos_rep = np.repeat([row[columnas_ultimas].values], 35, axis=0)
    df_temp = pd.DataFrame(datos_rep, columns=columnas_ultimas)
    df_nuevo_murcia2 = pd.concat([df_nuevo_murcia2, df_temp], ignore_index=True)



```

```python
df_super_murcia = pd.concat([df_nuevo_murcia, df_nuevo_murcia2], axis=1)
```

```python
import pandas as pd
import numpy as np

facultad_values = ['sociales+jur2015-2016', 'sociales+jur2016-2017', 'sociales+jur2017-2018', 'sociales+jur2018-2019', 'sociales+jur2019-2020', 'sociales+jur2020-2021', '-2021-2022.1', 'ing+arqui2015-2016', 'ing+arqui2016-2017', 'ing+arqui2017-2018', 'ing+arqui2018-2019', 'ing+arqui2019-2020', 'ing+arqui2020-2021', '-2021-2022.2', 'arte+hum2015-2016', 'arte+hum2016-2017', 'arte+hum2017-2018', 'arte+hum2018-2019', 'arte+hum2019-2020', 'arte+hum2020-2021', '-2021-2022.3', 'c.salud2015-2016', 'c.salud2016-2017', 'c.salud2017-2018', 'c.salud2018-2019', 'c.salud2019-2020', 'c.salud2020-2021', '-2021-2022.4', 'ciencias2015-2016', 'ciencias2016-2017', 'ciencias2017-2018', 'ciencias2018-2019', 'ciencias2019-2020', 'ciencias2020-2021', '-2021-2022.5']

df_super_murcia['facultad'] = pd.Series(np.tile(facultad_values, len(df_super_murcia) // len(facultad_values) + 1)[:len(df_super_murcia)])
```

```python

df_super_murcia.iloc[:, 1] = df_super_murcia.iloc[:, 1].replace('\.', np.nan, regex=True)

df_super_murcia.iloc[:, 1] = df_super_murcia.iloc[:, 1].fillna('2021-2022')
```

```python

df_super_murcia["facultad"].dtype
```

```python
for i in range(1, len(df_super_murcia)):
    if df_super_murcia.loc[i, "facultad"].startswith("-20"):
        df_super_murcia.loc[i, "facultad"] = df_super_murcia.loc[i-1, "facultad"]
```

```python
import re

df_super_murcia['facultad'] = df_super_murcia['facultad'].apply(lambda x: re.sub(r'\d+|-', '', x))
```

```python
facultad_dict = {
    'sociales+jur': 'Ciencias Sociales y Jurídicas',
    'ing+arqui': 'Ingeniería y Arquitectura',
    'arte+hum': 'Artes y Humanidades',
    'c.salud': 'Ciencias de la Salud',
    'ciencias': 'Ciencias'
}

df_super_murcia['facultad'] = df_super_murcia['facultad'].replace(facultad_dict)
```

```python
df_super_murcia = df_super_murcia.rename(columns={'Datos': 'matriculas', 'Año': 'año'})
```
