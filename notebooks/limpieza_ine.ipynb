```python
df_ine_murcia = pd.read_csv('ine_murcia.csv', sep=';', encoding='latin1')



df_ine_murcia.dtypes
                
```

```python
df_ine_murcia['Sexo'] = df_ine_murcia['Sexo'].astype(str)
df_ine_murcia['Provincias'] = df_ine_murcia['Provincias'].astype(str)
df_ine_murcia['Tasas'] = df_ine_murcia['Tasas'].astype(str)
df_ine_murcia['Periodo'] = df_ine_murcia['Periodo'].astype(str)
df_ine_murcia['Total'] = df_ine_murcia['Total'].str.replace(',', '.').astype(float)
```

```python
df_ine_murcia['Provincias'] = df_ine_murcia['Provincias'].str.extract(r'(\D+)', expand=False)
```

```python
df_ine_murcia = df_ine_murcia.drop('Provincias', axis=1)
```

```python
df_ine_murcia = df_ine_murcia[~df_ine_murcia['Periodo'].str.startswith(('200', '2011','2010'))].reset_index(drop=True)
```

```python
df_ine_murcia
```

```python
nuevos_indices = df_ine_murcia.index[::-1]

# Reindexa el DataFrame con los nuevos índices
df_ordenado = df_ine_murcia.reindex(nuevos_indices)
```

```python
df_ordenado = df_ordenado.reset_index(drop=True)
df_ordenado



df_ordenado = df_ordenado.iloc[:-2]
df_ordenado
```

```python
df_ordenado= df_ordenado.iloc[2:].reset_index(drop=True)
df_ordenado
```

```python
import pandas as pd

# Crear una lista para almacenar los diccionarios de datos
data = []

# Iterar sobre las filas del DataFrame en grupos de 4
for i in range(0, len(df_ordenado), 4):
    # Obtener los valores de la primera fila del grupo de 4
    sexo = df_ordenado.loc[i, 'Sexo']
    tasas = df_ordenado.loc[i, 'Tasas']
    periodo = df_ordenado.loc[i, 'Periodo']
    total_suma = df_ordenado.loc[i:i+3, 'Total'].sum()
    total_promedio = total_suma / 4
    
    # Crear un diccionario con los datos del grupo
    data.append({
        'Sexo': sexo,
        'Tasas': tasas,
        'Periodo': periodo,
        'Total': total_promedio
    })

# Crear el DataFrame final
df_promedio = pd.DataFrame(data)
```

```python
df_promedio 
```

```python
df_promedio['Periodo'] = df_promedio['Periodo'].str.replace(r'T[1-4]', '', regex=True)
```

```python
df_promedio
```

```python



df_promedio['Periodo'] = df_promedio['Periodo'].replace('2022', '2022-2023')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2021', '2021-2022')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2020', '2020-2021')

df_promedio['Periodo'] = df_promedio['Periodo'].replace('2019', '2019-2020')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2018', '2018-2019')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2017', '2017-2018')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2016', '2016-2017')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2015', '2015-2016')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2014', '2014-2015')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2013', '2013-2014')
df_promedio['Periodo'] = df_promedio['Periodo'].replace('2012', '2012-2013')


```

```python
# Cambiar el nombre del DataFrame
df_ine_murcia_limpio = df_promedio

# Guardar el DataFrame como archivo CSV
df_ine_murcia_limpio.to_csv('df_ine_murcia_limpio.csv', index=False)
```
